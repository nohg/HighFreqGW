```python
#Manual heterodying example
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

#time series
dt = 1/1000
ts = np.arange(0,1,dt)


#original sine wave sig
f1 = 100
sig1 = np.sin(2*np.pi*f1*ts)

#add some random noise
nse = 0.01 * np.random.random(size=len(ts))

sig1+= nse

plt.plot(ts,sig1)
```

```python
#second signal
f2 = 60
sig2 = np.sin(2*np.pi*f2*ts)
plt.plot(ts,sig2)
```

```python
#multiply to mix the signals
#should get signals with frequencies f1+f2 and f1-f2
mixedsig = sig1*sig2
plt.plot(ts,mixedsig)
```

```python
#make a psd with Welch's method

f , p = signal.welch(x = mixedsig, fs = 1/dt)
plt.semilogy(f, p)

#the frequencies f1-f2 and f1+f2 can be seen on the plot
```
